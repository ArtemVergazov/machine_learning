````python
def normalize(arr):
    """Shift by mean and scale by standard deviation each column of a matrix.

    Parameters
    ----------
    arr : array, shape = (n_rows, n_cols)
        The matrix, the columns of which to normalize.

    Returns
    ----------
    out : array, shape = (n_rows, n_cols)
        The final column-centered matrix.
    """
    assert arr.ndim == 2
    n_rows, n_cols = arr.shape

    ### BEGIN Solution (do not delete this comment)
    
    out = arr - arr.mean(axis=0)
    out /= out.std(axis=0)

    ### END Solution (do not delete this comment)

    return out
````

The main difference is a sun added outside the moons. In the cell below we generate the dataset.

````python
def make_sunny_moons(n_sun=50, n_moons=100, noise=0.0, sun_radius=1.9, theta=None):
    X_moons, y_moons = make_moons(n_samples=n_moons, noise=noise, random_state=0xC0FFEE)
    if not n_sun:
        return X_moons, y_moons
    
    np.random.seed(0xC0FFEE)
    angles = np.arange(0, 2 * np.pi, 2 * np.pi / n_sun)
    X_sun = sun_radius * np.column_stack([np.cos(angles), np.sin(angles)]) + np.array([0.5, 0.25])
    X_sun += np.random.normal(scale=noise, size=X_sun.shape)
    y_sun = 2 * np.ones(n_sun)

    X = np.vstack([X_moons, X_sun])
    y = np.concatenate([y_moons, y_sun]).astype(int)
    X -= X.mean(axis=0)

    if theta is None:
        theta = np.pi / 4
    c, s = np.cos(theta), np.sin(theta)
    R = np.array(((c,-s), (s, c)))
    X = X @ R
    
    return X, y

X, y = make_sunny_moons(n_sun=150, n_moons=300, noise=0.15)
````

### Subproblem 4.2. Pipeline: Fitting to Data by Feature Engineering (1 point)
In previous task 4.1 several classifiers obviously failed fitting to data. This happened because the decision rule of the classifier has a restricted form (e.g. linear for linear models), while the data is more complicated.

One may try to change the parameters of the classifier (e.g. increase the number of trees in Forest) in order to improve accuracy, but some models (especially linear) do not have parameters that can change the form of the decision rule.

In this case the **feature engineering** helps: one may try to compute new (e.g. non-linear) features based on the existing pool and fit the classifier in the new features. This may help low-complex classifiers to fit to hard data dependencies.

Your task it to

* `PolynomialFeatures`, `GaussianRandomProjection` among others - for feature generation
* `StandartScaler`, `MinMaxScaler` among others - for feature scaling
* `Pipeline` - for combining several operations in a row (e.g. feature creation & prediction)

````python
### BEGIN Solution (do not delete this comment)


def cartesian2polar(X):
    R = np.sqrt((X**2).sum(axis=1)).reshape((-1, 1))
    Phi = np.arctan2(X[:, 1], X[:, 0]).reshape((-1, 1))
    return np.hstack([R, Phi])


def polar2cartesian(X):
    X1 = (X[:, 0] * np.cos(X[:, 1])).reshape((-1, 1))
    X2 = (X[:, 0] * np.sin(X[:, 1])).reshape((-1, 1))
    return np.hstack([X1, X2])


rf_pipe = Pipeline([
    ('polar features', FunctionTransformer(func=cartesian2polar, inverse_func=polar2cartesian)),
    ('polynomial features', PolynomialFeatures(degree=4)),
    ('rf', RandomForestClassifier(n_estimators=20, max_depth=4, n_jobs=-1, random_state=0xC0FFEE)),
])

lr_pipe = Pipeline([
    ('generate features', PolynomialFeatures(degree=4)),
    ('lr', LogisticRegression(random_state=0xC0FFEE, n_jobs=-1)),
])

_, axes = plt.subplots(ncols=2, nrows=1, figsize=(20,10))

rf_pipe.fit(X, y)
plot_decision_regions(X, y, rf_pipe, ax=axes[0], legend=2, scatter_kwargs={'s': 30})
accuracy = accuracy_score(y, rf_pipe.predict(X))
axes[0].set_title(f'RandomForest with GaussianRandomProjection, accuracy = {round(accuracy, 2)}')
axes[0].tick_params(axis='both', which='major', labelsize=8)

lr_pipe.fit(X, y)
plot_decision_regions(X, y, lr_pipe, ax=axes[1], legend=2, scatter_kwargs={'s': 30})
accuracy = accuracy_score(y, lr_pipe.predict(X))
axes[1].set_title(f'LogisticRegression with PolynomialFeatures, accuracy = {round(accuracy, 2)}')
axes[1].tick_params(axis='both', which='major', labelsize=8)

plt.show()

### END Solution (do not delete this comment)
````

**Your text answer (do not delete this comment)**:

First I tried to apply usual PolynomialFeatures to RandomForest classifier, but the accuracy only decreased. So I came up with another transformation. RandomForest classifier applies decision rules based on $x$ and $y$ coordinates because our initial dataset is represented in cartesian coordinates. However, it can be seen that the moons follow a specific sphere-like pattern which can be expressed in polar coordinates more naturally than in cartesian. That's why I applied FunctionTransformer which transforms cartesian coordinates to polar coordinates. However, the accuracy in polar coordinates still was unsatisfactory because even in polar coordinates the pattern of the moons placement is quite complicated.
````python
### BEGIN Solution (do not delete this comment)

# Important hyperparameters: 
# - max depth
# - n_estimators
# - max_features


def forest_grid_search(forest_constructor, max_depth_range, n_estimators_range, max_features_range):
    best_max_depth = max_depth_range[0]
    best_n_estimators = n_estimators_range[0]
    best_max_features = max_features_range[0]
    
    best_score = np.inf
    
    for max_depth in max_depth_range:
        for n_estimators in n_estimators_range:
            for max_features in max_features_range:
                forest = forest_constructor(
                    n_estimators=n_estimators,
                    max_depth=max_depth,
                    max_features=max_features,
                    n_jobs=-1,
                    random_state=0xC0FFEE,
                )
                forest.fit(X_train, y_train)

                y_pred = forest.predict(X_val)
                score = mse(y_val, y_pred)

                if score < best_score:
                    best_score = score
                    
                    best_forest = forest
                    best_max_depth = max_depth
                    best_n_estimators = n_estimators
                    best_max_features = max_features
    return best_max_depth, best_n_estimators, best_max_features


n_features = X.shape[1]

max_depth_range = range(2, 20, 2)
n_estimators_range = range(10, 150, 20)
max_features_range = ['sqrt', n_features // 2, n_features]

best_max_depth, best_n_estimators, best_max_features = forest_grid_search(RandomForestRegressor,
                                                                          max_depth_range,
                                                                          n_estimators_range,
                                                                          max_features_range)
random_forest = RandomForestRegressor(
    n_estimators=best_n_estimators,
    max_depth=best_max_depth,
    max_features=best_max_features,
    n_jobs=-1,
    random_state=0xC0FFEE,
)

print('Best parameters for RandomForestRegressor:')
print(f'n_estimators:\t{best_n_estimators}')
print(f'max_depth:\t{best_max_depth}')
print(f'max_features:\t{best_max_features}')
print()

random_forest.fit(X_train, y_train)

print('RandomForestRegressor scores:')
print(f'train mse:\t{mse(y_train, random_forest.predict(X_train))}')
print(f'val mse:\t{mse(y_val, random_forest.predict(X_val))}')
print(f'test mse:\t{mse(y_test, random_forest.predict(X_test))}')
print()

best_max_depth, best_n_estimators, best_max_features = forest_grid_search(ExtraTreesRegressor,
                                                                          max_depth_range,
                                                                          n_estimators_range,
                                                                          max_features_range)
extra_trees = ExtraTreesRegressor(
    n_estimators=best_n_estimators,
    max_depth=best_max_depth,
    max_features=best_max_features,
    n_jobs=-1,
    random_state=0xC0FFEE,
)

print('Best parameters for ExtraTreesRegressor:')
print(f'n_estimators:\t{best_n_estimators}')
print(f'max_depth:\t{best_max_depth}')
print(f'max_features:\t{best_max_features}')
print()

extra_trees.fit(X_train, y_train)

print('ExtraTreesRegressor scores:')
print(f'train mse:\t{mse(y_train, extra_trees.predict(X_train))}')
print(f'val mse:\t{mse(y_val, extra_trees.predict(X_val))}')
print(f'test mse:\t{mse(y_test, extra_trees.predict(X_test))}')

### END Solution (do not delete this comment)
````

### Subproblem 6.1. AdaBoost alphas and distributions (1 point)

Now let us define functions to calculate alphas and distributions for AdaBoost algorithm

````python
def ada_boost_alpha(y, y_pred_t, distribution):
    """
    Function, which calculates the weights of the linear combination of the classifiers.
    
    y_pred_t is a prediction of the t-th base classifier
    """
    ### BEGIN Solution (do not delete this comment)

    N = sum(distribution * (y * y_pred_t <= 0))
    alpha = .5 * np.log((1 - N) / N)
    
    ### END Solution (do not delete this comment)

    return alpha

def ada_boost_distribution(y, y_pred_t, distribution, alpha_t):
    """
    Function, which calculates sample weights
    
    y_pred_t is a prediction of the t-th base classifier
    """
    ### BEGIN Solution (do not delete this comment)

    w = distribution * np.exp(-alpha_t * y_pred_t * y)
    distribution = w / sum(w)
    
    ### END Solution (do not delete this comment)

    return distribution
````

````python
class Boosting():
    """
    Generic class for construction of boosting models
    
    :param n_estimators: int, number of estimators (number of boosting rounds)
    :param base_classifier: callable, a function that creates a weak estimator. Weak estimator should support sample_weight argument
    :param get_alpha: callable, a function, that calculates new alpha given current distribution, prediction of the t-th base estimator,
                      boosting prediction at step (t-1) and actual labels
    :param get_distribution: callable, a function, that calculates samples weights given current distribution, prediction, alphas and actual labels
    """
    def __init__(self, n_estimators=50, base_classifier=None,
                 get_alpha=ada_boost_alpha, update_distribution=ada_boost_distribution):
        self.n_estimators = n_estimators
        self.base_classifier = base_classifier
        self.get_alpha = get_alpha
        self.update_distribution = update_distribution
        
    def fit(self, X, y):
        n_samples = len(X)
        distribution = np.ones(n_samples, dtype=float) / n_samples
        self.classifiers = []
        self.alphas = []
        for i in range(self.n_estimators):
            # create a new classifier
            self.classifiers.append(self.base_classifier())     
            self.classifiers[-1].fit(X, y, sample_weight=distribution)

            ### BEGIN Solution (do not delete this comment)
            
            # make a prediction
            y_pred = self.classifiers[-1].predict(X)
            
            # update alphas, append new alpha to self.alphas
            alphas = self.get_alpha(y, y_pred, distribution)
            self.alphas.append(alphas)
            
            # update distribution and normalize
            distribution = self.update_distribution(y, y_pred, distribution, alphas)
            
            ### END Solution (do not delete this comment)

    
    def predict(self, X):
        final_predictions = np.zeros(X.shape[0])
    
        ### BEGIN Solution (do not delete this comment)
        
        #get the weighted votes of the classifiers
        final_predictions = sum(alpha * clf.predict(X) for alpha, clf in zip(self.alphas, self.classifiers))
        out = np.sign(final_predictions)
        
        ### END Solution (do not delete this comment)

        return out
````
